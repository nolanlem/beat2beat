```python
def map2drum(midinote):
    weirdoinstruments = range(60,82)
    if np.isin(midinote, [35,36]):
        note = 1 # bass drum
    if np.isin(midinote,[37]):
        note = 2 # side stick
    if np.isin(midinote, [38,39,40]):
        note = 3 # snares
    if np.isin(midinote, [41, 45, 47]):
        note = 4 # low toms
    if np.isin(midinote, [43, 48, 50]):
        note = 5 # high tom
    if np.isin(midinote, [42, 44, 46]):
        note = 6 # hh 
    if np.isin(midinote, [51, 53 , 54, 56, 59]):
        note = 7 # ride
    if np.isin(midinote, [49, 52, 55, 57, 58]):
        note = 8 # crash
    if np.isin(midinote, weirdoinstruments):
        note = 9 # other
    return note

def drum2MIDI(drumnote):
    if drumnote == 0:
        midinote = 60
    if drumnote == 1:
        midinote = 35
    if drumnote == 2:
        midinote = 37
    if drumnote == 3:
        midinote = 41
    if drumnote == 4:
        midinote = 41 
    if drumnote == 5:
        midinote = 43 
    if drumnote == 6:
        midinote = 42
    if drumnote == 7:
        midinote=51 
    if drumnote == 8:
        midinote = 49
    if drumnote == 9:
        midinote == 60
    return midinote
    
```
